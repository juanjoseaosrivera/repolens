```python
import numpy as np

from repolens.embeddings import Embedder

texts = [
    "Picasso",
    "Light year",
    "painting",
    "Horse power",
]
vectors = np.array(Embedder().embed(texts))


def cosine(a, b):
    return float((a @ b) / (np.linalg.norm(a) * np.linalg.norm(b)))


for i, t1 in enumerate(texts):
    for j, t2 in enumerate(texts):
        if i < j:
            print(f"{cosine(vectors[i], vectors[j]):.3f}  {t1!r}  ↔  {t2!r}")
```
